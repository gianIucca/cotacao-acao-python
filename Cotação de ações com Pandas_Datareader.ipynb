## Programa simples em Python para buscar o valor de ações e mostrar elas graficamente utilizando as bibliotecas Pandas, Plotly e Pandas_DataReader

```python
from pandas_datareader import data as web
import pandas as pd 
import matplotlib.pyplot as plt 
import plotly.express as px
```

```python
lista_empresas = ['ITUB4', 'PETR4', 'BBDC4', 'VALE3']
data_inicial = '01-01-2021'
data_final = '01-01-2022'

for empresa in lista_empresas:
    print(f'{empresa}:')
    df = web.DataReader(f'{empresa}.SA', data_source='yahoo', start=data_inicial, end=data_final)
    fig = px.bar(df, y='Adj Close')
    fig.show()
    df['Adj Close'].plot(figsize=(10,10))
    plt.show()
```
